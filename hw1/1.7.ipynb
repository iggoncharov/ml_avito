```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# график 1
s1 = np.random.multivariate_normal(mean = [20, 10], cov = [[1,0.7],[0.7,1]], size = 1000)

# график 2
s2 = np.random.geometric(p=0.1, size = 3000)

# график 3
s3 = np.linspace(0.01, 6, 201)

# график 4
s4 = np.random.normal(loc = 5, size = 14)
date = [str(i) for i in np.datetime64('2021-01-01') + np.arange(0, 14)]
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (15, 8))

# график 1
ax[0, 0].scatter(s1[:, 0], s1[:, 1], s = 50, edgecolors = 'black', linewidth = 1)
ax[0, 0].set_title('Scatter plot example')

# график 2
ax[0, 1].hist(s2, bins = 40, color = 'red')
ax[0, 1].set_title('Hist plot example')

# график 3
ax[1, 0].plot(s3, np.log(s3), color = 'black', label = 'log(x)')
ax[1, 0].plot(s3, np.log(2*s3), color = 'red', label = 'log(2x)')
ax[1, 0].set_title('Line plot example')
ax[1, 0].legend()

# график 4
ax[1, 1].bar(date, s4)
ax[1, 1].tick_params(axis='x', rotation = 90)
ax[1, 1].set_title('Bar plot example')


fig.suptitle('Plot Examples', x = 0.5, fontsize = 20, y = 0.95)
fig.set(facecolor = 'white') #фон
plt.subplots_adjust(wspace=0.1, hspace=0.2)
plt.show()

#fig.savefig('saved_figure.png') # <- сохранить в png
```
